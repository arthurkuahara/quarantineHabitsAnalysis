### Twitter - Análise de Sentimentos (TSA)
A mineração de opinião e sentimento é uma importante área de pesquisa porque, devido ao grande número de postagens diárias nas redes sociais, extrair a opinião das pessoas é uma tarefa desafiadora. A análise de sentimento tem muitas aplicações em diversas áreas. 

A escolha de desenvolver uma análise de sentimento sobre Tweets postados no período da quarentena é de grande relação com os dados analisados nos tópicos anteriores. Visto que o objetivo trata-se de analisar a mudança de hábitos e comportamento das pessoas durante a pandemia, nada mais coerente que analisar também seus sentimentos e reações nesse período.

O Twitter é um rico recurso para extração de sentimentos, visto que possuem um alto grau de subjetividade dos pensamentos e sua análise de texto é mais simples com o restrito número de caracteres de cada tweet. Além disso, entende-se que o twitter seja uma das redes sociais mais famosas e utilizadas - com mais de 300 milhões de usuários ativos - e é ideal para uma análise dos sentimentos dos usuários em rede global.

Nesse trabalho, desenvolvemos a polaridade dos tweets baseados nos tópicos e hashtags para caracterizar o tema do Covid19.

#### O que é a análise de sentimentos?
Análise de sentimento é um processo automatizado de identificar e classificar informações subjetivas de um dado de texto. Pode se tratar de opiniões, julgamentos ou sensações sobre um particular tópico ou feature. No nosso caso, temos como feature os tópicos relacionados ao Corona Vírus.
A forma mais comum de análise de sentimento é chamada de **detecção de polaridade**, que será a forma utilizada nesse trabalho, e envolve classificar um texto em: positivo, negativo ou neutro.

## Regressão
Para a regressão, enfrentamos a dificuldade de, após termos criado todo o dataset com a categorização para tweets positivos, negativos ou neutros, notamos que precisaríamos também de um valor real, como a polaridade não categorizada para que fosse possível aplicarmos o modelo de regressão. Dessa forma, outro notebook para definição dos dados foi criado: Twitter_Get_Polarity onde guardamos em um novo csv o valor real para a polaridade de cada tweet.

Por essa dificuldade, é utilizado de um dataset menor que o utilizado para a classificação, visto que o TextBlob demora um tempo considerável para a análise de sentimento de cada tweet e faltou o tempo disponível para o notebook rodar por longas horas. Apesar dessa consideração, isso não é um problema, visto que ainda assim temos uma grande amostra com aproximadamente 12k tweets.

Os modelos a seguir são formulados pela variável de texto e tentam prever o valor da polaridade.

### Primeira tentativa: Linear Regression

O primeiro modelo utilizado foi Linear Regression. Note porém que a modelagem dos nossos dados não possui um comportamento linear e aplicá-los sob esse modelo linear gerou um valor de R2 negativo. Isto significa que estávamos fazendo pior que tirar o valor médio. Por isso, após verificarmos esse resultado, decidimos verificar outros modelos para encontrarmos o melhor valor para o R2 score.

### Segunda tentativa: Rigde Regression

O segundo modelo escolhido foi Rigde Regression. Nesse modelo, obtivemos um melhor valor para o r2 score, de 0.34. Por estar acima de 0, quer dizer que estamos fazendo um pouco melhor que a média, mas ainda não é uma pontuação satisfatória. Apesar disso, somos capazes de justificar e apontar várias observações para que fossémos capazes de "tunar" o modelo e conseguir melhores resultados.

Primeiramente, é importante observar que os nossos dados coletas foram "puros", diretamente realistas. Isto quer dizer que todos os valores que temos de tweet e inclusive de polaridade foram coletados por nós e todo o trabalho de pre-processamento feito sobre os tweets foi feito por nós. Assim, diferentemente de dados já limpos disponíveis em outras plataformas, como Kaggle, temos que lidar com dados que ainda precisariam de melhores processamentos para tirar qualquer tipo de ruído que possa causar sobre o modelo. 

Algumas melhorias de limpeza foram feitas nas fases anteriores, mas ainda assim poderíamos propor algumas outras melhor, como "lemmatizar" que é capaz de simplificar as palavras usadas nos tweets. Isso não foi implementado nesse trabalho, por fugir do escopo e demandar mais algumas análises exploratórias, porém acreditamos que o modelo possa ser devidamente melhorado com esse tratamento prévio dos tweets.

#### Coeficientes de Ridge Plot como uma função da regularização
Ridge Regression é o estimador usado na regressão. Cada cor representa um recurso diferente do vetor de coeficientes e isso é exibido como uma função do parâmetro de regularização.

Podemos notar a utilidade de aplicar a regressão de Ridge a matrizes altamente mal condicionadas, como acontece no nosso caso. Para tais matrizes, uma ligeira mudança na variável de destino pode causar grandes variações nos pesos calculados. Nesses casos, é útil definir uma certa regularização (alfa) para reduzir essa variação (ruído).

Quando alfa é muito grande, o efeito de regularização domina a função de perda ao quadrado e os coeficientes tendem a zero. No final do caminho, como alfa tende para zero e a solução tende para os mínimos quadrados ordinários, os coeficientes exibem grandes oscilações. Na prática, é necessário sintonizar o alfa de forma que haja equilíbrio entre os dois.
